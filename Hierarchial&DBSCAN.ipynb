```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_moons, make_circles
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

```python
x_blob , y_blob = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)
x_blob
```

```python
plt.scatter(x_blob[:,0], x_blob[:,1])
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x_blob)
linkage = linkage(x_scaled, method='ward')
```

```python
plt.figure(figsize=(10,5))
dendrogram(linkage, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Dendrogram')
plt.xlabel('Cluster')
plt.ylabel('Distance')
plt.show()
```

```python
hc_clusters = fcluster(linkage, t=4, criterion='maxclust')
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=x_blob[:,0],y=x_blob[:,1], hue=hc_clusters, palette='tab10')
plt.title("Hierarchical Clustering")
plt.xlabel("X1")
plt.ylabel("X2")
```

```python
x_moon, y_moon = make_moons(n_samples=200, noise=0.05, random_state=42)
plt.scatter(x_moon[:,0], x_moon[:,1])
plt.xlabel('X1')
plt.ylabel('X2')
```

```python
x_moon_scaled = scaler.fit_transform(x_moon)
```

```python
db = DBSCAN(eps=0.3, min_samples=5)
db_labels = db.fit_predict(x_moon_scaled)
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=x_moon[:,0],y=x_moon[:,1], hue=db_labels, palette='tab10')
plt.title("DBSCAN")
plt.xlabel("X1")
plt.ylabel("X2")
```

```python
x_circle, y_circle = make_circles(n_samples=300, noise=.1, random_state=42)
plt.figure(figsize=(10,5))
plt.scatter(x_circle[:,0], x_circle[:,1])
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
Xc_s = scaler.fit_transform(x_circle)
```

```python
db = DBSCAN(eps=0.3, min_samples=15)
db_labels = db.fit_predict(Xc_s)
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=x_circle[:,0],y=x_circle[:,1], hue=db_labels, palette='tab10', legend="auto")
plt.title("DBSCAN")
plt.xlabel("X1")
plt.ylabel("X2")
```

## Project Summary

This notebook provides an exploration of different clustering algorithms on synthetic datasets.
